Load the libaries & Data

```python
import pandas as pd
import numpy as np
np.set_printoptions(suppress=True)

all_data=pd.read_csv('opp.csv')
all_data.head()
```

```python
all_data=all_data.dropna()
all_data=all_data.drop(['case','Pop','sex','site'],axis=1)
all_data.head()
```

```python
train_data=all_data.head(52)
test_data=all_data.head(-52)
```

```python
from statsmodels.nonparametric.kernel_regression import KernelReg
y=train_data['age'].values
x1=train_data['hdlngth'].values
x2=train_data['skullw'].values
x3=train_data['totlngth'].values
x4=train_data['taill'].values
x5=train_data['footlgth'].values
x6=train_data['earconch'].values
x7=train_data['eye'].values
x8=train_data['chest'].values
x9=train_data['belly'].values
model = KernelReg(endog=y, exog= [x1,x2,x3,x4,x5,x6,x7,x8,x9], var_type='ccccccccc')
```

```python
x1=test_data['hdlngth'].values
x2=test_data['skullw'].values
x3=test_data['totlngth'].values
x4=test_data['taill'].values
x5=test_data['footlgth'].values
x6=test_data['earconch'].values
x7=test_data['eye'].values
x8=test_data['chest'].values
x9=test_data['belly'].values
preds_model=model.fit([x1,x2,x3,x4,x5,x6,x7,x8,x9])
preds=np.round(preds_model[0])
actuals=test_data['age']

data={'actuals': actuals,
     'preds': preds}
comparison=pd.DataFrame(data)
comparison.head()
```
